```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import json
import numpy as np
import seaborn as sns
```

```python
df_tweet_2 = df
```

```python
# Aggiungo uno spazio prima e dopo ogni stopwords in modo tale 
# da non coniderare il caso in cui la stopword è contenuta in una parola
words = set(stopwords.words('italian'))
stopwords = [' ' + x + ' ' for x in words]
```

```python
emoticons =  ('😇','😊','❤️','😘','💞','💖','🤗','💕','👏','🎉','👍',
              '😂','😡','😠','😭','🤦‍','🤷🏼‍','😞','👎','😱','😓','🔝')
```

```python
df_tweet_2.testo = df_tweet_2.testo.replace("@[\w]*[_-]*[\w]*"," ",regex=True)   # rimuovo i tag
```

```python
df_tweet_2.testo = df_tweet_2.testo.replace("https?://[\w/%-.]*"," ",regex=True) # rimuovo i link url
```

```python
# Rimuovo tutto tranne le lettere dell'alfabeto e le emoticons
df_tweet_2.testo = df_tweet_2.testo.replace('[^ a-zA-Zà-ú'
                            '\😇\😊\❤️\😘\💞\💖\🤗\💕\👏\🎉\👍'
                            '\😂\😡\😠\😭\🤦‍\🤷🏼‍\😞\😱\👎\😓\🔝]', " ",regex=True)    
for word in emoticons:
    df_tweet_2.testo = df_tweet_2.testo.replace(word, " "+word+" ",regex=True) 

df_tweet_2.testo = df_tweet_2.testo.replace('\s+', ' ',regex=True)               # Rimuovo gli spazi in eccesso 
df_tweet_2.testo = df_tweet_2.testo.replace('^ ', '', regex=True)                # Rimuovo lo spazio all'inizio
df_tweet_2.testo = df_tweet_2.testo.replace(' $', '', regex=True)                # Rimuovo lo spazio alla fine
df_tweet_2.testo = df_tweet_2.testo.apply(lambda x: x.lower())                   # Rendo tutto minuscolo
```

```python
for word in stopwords:
    df_tweet_2.testo = df_tweet_2.testo.replace(word, ' ',regex=True)
```

```python
# Rimuovo gli spazi all'inizio e alla fine di ogni tweet
df_tweet_2.testo = df_tweet_2.testo.apply(lambda x: x.strip())     
```

```python
# Rimuovo i tweet rimasti vuoti
df_tweet_2 = df_tweet_2[df_tweet_2.testo != '']                  
```
