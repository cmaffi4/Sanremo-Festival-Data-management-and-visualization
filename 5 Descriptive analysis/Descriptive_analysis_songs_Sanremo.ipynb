```python
df_valori_medi = df.groupby('Anno').agg({"acousticness":"mean",  "danceability" : "mean", "duration_ms" : "mean" , 
                        "energy" : "mean" ,   "instrumentalness" : "mean",  "key" : "mean" ,  "liveness" : "mean" ,
                         "loudness" : "mean" , "mode" : "mean", "speechiness" : "mean", "tempo" : "mean" , 
                        "time_signature" : "mean",  "valence" : "mean"})
df_valori_medi
```

```python
df_vincitori = df[df['vincitore']==1]
df_vincitori
```

```python
df_confronto = pd.merge(df_vincitori, df_valori_medi, on='Anno', suffixes=('_win', '_mean'))
df_confronto
```
