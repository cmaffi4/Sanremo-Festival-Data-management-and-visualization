```python
# Setup
from kafka import KafkaConsumer
import json
from pprint import pprint
from pymongo import MongoClient
```

```python
def check_tabulation_and_newline(data):
    if type(data) == str:
        return data.replace("\t", " ").replace("\n", " ")
    else:
        return data
```

```python


myclient = MongoClient('localhost', 27017)
mydb = myclient["Sanremo_project"]
mycol = mydb["Sanremo_tweets"]


consumer = KafkaConsumer(
  bootstrap_servers=["kafka:9092"],
  auto_offset_reset="latest",
  value_deserializer=lambda v: json.loads(v.decode("utf-8")))


consumer.subscribe(["sanremotweet"])

try:
    for msg in consumer:
        data = msg.value
        data = {key:check_tabulation_and_newline(data[key]) for key in data}
        mycol.insert_one(data)
except:
    print("An exception occurred")
    

```
